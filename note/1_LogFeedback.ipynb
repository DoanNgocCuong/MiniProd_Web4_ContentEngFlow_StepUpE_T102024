# 0. Cấu trúc code hiện tại 

Được rồi, dưới đây là bảng tóm tắt chi tiết về **Routes**, **Controllers**, và **Services** trong ứng dụng web, được giải thích đơn giản cho học sinh cấp 2:

---

### **Bảng tóm tắt**

| **Thành phần**     | **Ví dụ đời thường**                                       | **Vai trò trong ứng dụng web**                                                          |
|--------------------|------------------------------------------------------------|-----------------------------------------------------------------------------------------|
| **Routes (Đường dẫn)**       | **Cửa ra vào của nhà hàng**                                  | - Định nghĩa các địa chỉ (URL) mà người dùng truy cập.<br>- Chuyển yêu cầu của người dùng đến **Controller** phù hợp. |
| **Controller (Người phục vụ)** | **Nhân viên phục vụ nhận gọi món từ khách hàng**              | - Nhận yêu cầu từ người dùng qua **Routes**.<br>- Hiểu người dùng muốn gì và quyết định cần làm gì.<br>- Gọi **Service** để thực hiện công việc chính.<br>- Trả kết quả lại cho người dùng. |
| **Service (Nhà bếp)**        | **Đầu bếp trong nhà bếp nấu món ăn theo yêu cầu**             | - Thực hiện công việc chính (xử lý dữ liệu, tính toán).<br>- Tương tác với cơ sở dữ liệu hoặc dịch vụ bên ngoài.<br>- Trả kết quả về cho **Controller** sau khi hoàn thành. |

````javascript:backend/src/services/larkbaseService.js
const axios = require('axios');

const LARKBASE_CONFIG = {
    app_id: "cli_a7852e8dc6fc5010",
    app_secret: "6SIj0RfQ0ZwROvUhkjAwLebhLfJkIwnT", 
    app_base_token: "FjRbbDy10aGpKfso9uxl646Gguc",
    base_table_id: "tble369ZpDcMr2fu"
};

class LarkbaseService {
    static async createRecords(data) {
        const payload = {
            config: LARKBASE_CONFIG,
            records: [{
                fields: {
                    feedback: data
                }
            }]
        };

        return await axios.post(
            'http://103.253.20.13:25033/api/larkbase/create-many-records',
            payload,
            {
                headers: {
                    'Content-Type': 'application/json'
                }
            }
        );
    }
}

module.exports = LarkbaseService;
````

````javascript:backend/src/controllers/feedbackController.js
const LarkbaseService = require('../services/larkbaseService');

exports.submitFeedback = async (req, res) => {
    try {
        const { feedback } = req.body;

        if (!feedback || typeof feedback !== 'string') {
            return res.status(400).json({
                error: 'Invalid feedback format. Feedback must be a non-empty string.'
            });
        }

        const response = await LarkbaseService.createRecords(feedback);

        console.log('Feedback submitted successfully:', {
            feedback,
            response: response.data
        });

        res.json({
            message: 'Feedback submitted successfully',
            data: response.data
        });

    } catch (error) {
        console.error('Error submitting feedback:', {
            message: error.message,
            stack: error.stack
        });

        res.status(500).json({
            error: 'Failed to submit feedback',
            details: error.message
        });
    }
};
````

Key changes made:

1. Created a new `LarkbaseService` class in a separate file to handle all Larkbase-related operations
2. Moved the Larkbase configuration to a constant in the service file
3. Created a static `createRecords` method to handle the API call
4. Simplified the controller to focus on request handling and response formatting
5. Used dependency injection pattern by requiring the service in the controller

This separation of concerns makes the code:
- More maintainable
- Easier to test
- More reusable
- Easier to modify Larkbase-related functionality without touching the controller

Remember to create the `services` directory in your backend structure if it doesn't exist yet.
