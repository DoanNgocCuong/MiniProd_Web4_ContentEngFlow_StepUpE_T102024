

Điều này xảy ra vì bạn đang sử dụng `node` thay vì `nodemon` khi chạy `npm start`. 

Trong `package.json` của bạn có 2 scripts:
```json
"scripts": {
    "start": "node src/server.js",    // Không tự động restart
    "dev": "nodemon src/server.js"    // Tự động restart khi có thay đổi
}
```

1. `npm start` sử dụng `node` để chạy server:
   - Không theo dõi thay đổi file
   - Cần phải dừng và chạy lại server thủ công khi có thay đổi code

2. `npm run dev` sử dụng `nodemon`:
   - Tự động theo dõi thay đổi file
   - Tự động restart server khi phát hiện thay đổi
   - Phù hợp cho môi trường development

Để fix vấn đề này, bạn nên:

1. Trong quá trình development, sử dụng:
```bash
npm run dev
```

2. Hoặc sửa script trong package.json:
```json
{
  "scripts": {
    "start": "nodemon src/server.js",  // Thay node bằng nodemon
    "dev": "nodemon src/server.js"
  }
}
```

3. Hoặc tạo thêm script riêng cho production:
```json
{
  "scripts": {
    "start": "node src/server.js",         // For production
    "dev": "nodemon src/server.js",        // For development
    "prod": "NODE_ENV=production node src/server.js"  // Explicit production
  }
}
```

Lời khuyên:
- Trong môi trường development: sử dụng `npm run dev`
- Trong môi trường production: sử dụng `npm start`
- Không nên sử dụng `nodemon` trong production vì nó tiêu tốn tài nguyên không cần thiết
