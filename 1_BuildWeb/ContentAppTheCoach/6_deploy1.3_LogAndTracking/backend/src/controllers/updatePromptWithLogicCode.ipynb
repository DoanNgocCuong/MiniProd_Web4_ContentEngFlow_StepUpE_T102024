Giải thích:
1. Code sẽ xử lý từng lesson trong mảng lessons
2. Với mỗi lesson:
   - Lấy structure và 3 phrases (main + 2 optional)
   - Tạo 3 câu bằng cách thay thế "_______" trong structure với mỗi phrase
   - Với mỗi câu, tạo phiên bản sentence_hide bằng cách thay thế mọi ký tự (trừ khoảng trắng) bằng dấu "_"
3. Kết quả cuối cùng là một mảng chứa tất cả các câu được tạo ra (trong ví dụ này là 6 câu từ 2 lessons)

- description: Luôn là "TRẢ LỜI THEO TRANH GỢI Ý"
- sentence_en: Câu tiếng Anh hoàn chỉnh
- sentence_hide: Câu với các ký tự được thay thế bằng dấu gạch dưới
